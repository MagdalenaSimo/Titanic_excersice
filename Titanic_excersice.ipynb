```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
titanic_survived = pd.read_csv('gender_submission.csv')
titanic_survived
```

```python
titanic_passenger = pd.read_csv('test.csv')
titanic_passenger
```

```python
#merged two tables together
titanic_train = pd.merge(titanic_passenger,titanic_survived,how='inner',on='PassengerId')
titanic_train
```

```python
#removed the tickets and cabin atribute

titanic_train.drop(columns=['Ticket','Cabin'])
```

```python
#find the null values in dataframe
titanic_train.isnull().sum()
```

```python
# Replace using median in column Age
median = titanic_train['Age'].median()
titanic_train['Age'].fillna(median, inplace=True)
titanic_train
```

```python
#replace the null values in column Fare
median2 = titanic_train['Fare'].median()
titanic_train['Fare'].fillna(median2, inplace=True)
titanic_train
```

```python
#how many null values is in the dataframe
titanic_train.isnull().sum()
```

```python
#remove the null values in cabin 
titanic_train.dropna(inplace=True)
titanic_train
```

```python
#find the null values in dataframe
titanic_train.isnull().sum()
```
